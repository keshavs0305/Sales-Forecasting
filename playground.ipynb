```python
def itemwise_freq(cus_data):
    freq_itemwise = {}
    x = cus_data['Order date']
    for i in x.index:
        freq_itemwise[i] = x[i]/x.sum()
    return freq_itemwise
```

```python
def num_days_since_last(itemwise_data):
    last_buy = {}
    for item_id in cus_data['Item Id'].unique():
        x = itemwise_data.get_group(item_id).reset_index()
        if len(x) > 1:
            last_buy[item_id] = (x.iloc[len(x)-1]['Order date'] - x.iloc[len(x)-2]['Order date']).days
    return last_buy
```

```python
def data_till_date(cus_id):
    cus_data = df.groupby('Custumer Id').get_group(cus_id)
    number_of_orders = len(cus_data.groupby('Order date').count())
    itemwise_data = itemwise_freq(cus_data.groupby('Item Id').count())
    last_ord = num_days_since_last(cus_data.groupby('Item Id'))
```
